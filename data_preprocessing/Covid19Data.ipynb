```python
# Load libraries
import pandas as pd
import seaborn as sns
```

```python
# Load county dataset
county_data = pd.read_csv('us_counties_covid19_daily.csv')
county_data = county_data.fillna(0)
print(len(county_data))
county_data.head()
```

```python
# Load state dataset
state_data = pd.read_csv('us_states_covid19_daily.csv')
print(len(state_data))

state_data["date"] = pd.to_datetime(state_data["date"], format='%Y%m%d')
state_data = state_data.sort_values('date')
data_to_plot = state_data.loc[state_data["state"] == "AL", "positiveIncrease"]
print(len(data_to_plot))
data_to_plot.head()
```

```python
# Clean state dataset
state_data = state_data[['date', 'state', 'fips', 'positive', 'negative', 'death', 'totalTestResults', 'positiveIncrease', 'negativeIncrease', 'deathIncrease', 'totalTestResultsIncrease']]
state_data = state_data.fillna(0)
state_data.head()
```

```python
# Load country dataset
us_data = pd.read_csv('us_covid19_daily.csv')
print(len(us_data))
us_data.head()
```

```python
# Clean country data
us_data = us_data[['date', 'states', 'positive', 'negative', 'hospitalizedCurrently', 'death', 'totalTestResults', 'positiveIncrease', 'negativeIncrease', 'hospitalizedIncrease','deathIncrease','totalTestResultsIncrease']]
us_data = us_data.fillna(0)
us_data.head()
```

```python
# Load vaccine dataset
vaccine_data_state = pd.read_csv('us_state_vaccinations.csv')
print(len(vaccine_data_state))
vaccine_data_state.head()
```

```python
# Clean vaccine dataset
vaccine_data_state = vaccine_data_state[['date', 'location', 'total_vaccinations', 'total_distributed', 'people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations', 'share_doses_used']]
vaccine_data_state.head()
```

```python
# map state name to state code
us_state_abbrev = {
    'Alabama': 'AL',
    'Alaska': 'AK',
    'American Samoa': 'AS',
    'Arizona': 'AZ',
    'Arkansas': 'AR',
    'California': 'CA',
    'Colorado': 'CO',
    'Connecticut': 'CT',
    'Delaware': 'DE',
    'District of Columbia': 'DC',
    'Florida': 'FL',
    'Georgia': 'GA',
    'Guam': 'GU',
    'Hawaii': 'HI',
    'Idaho': 'ID',
    'Illinois': 'IL',
    'Indiana': 'IN',
    'Iowa': 'IA',
    'Kansas': 'KS',
    'Kentucky': 'KY',
    'Louisiana': 'LA',
    'Maine': 'ME',
    'Maryland': 'MD',
    'Massachusetts': 'MA',
    'Michigan': 'MI',
    'Minnesota': 'MN',
    'Mississippi': 'MS',
    'Missouri': 'MO',
    'Montana': 'MT',
    'Nebraska': 'NE',
    'Nevada': 'NV',
    'New Hampshire': 'NH',
    'New Jersey': 'NJ',
    'New Mexico': 'NM',
    'New York': 'NY',
    'North Carolina': 'NC',
    'North Dakota': 'ND',
    'Northern Mariana Islands':'MP',
    'Ohio': 'OH',
    'Oklahoma': 'OK',
    'Oregon': 'OR',
    'Pennsylvania': 'PA',
    'Puerto Rico': 'PR',
    'Rhode Island': 'RI',
    'South Carolina': 'SC',
    'South Dakota': 'SD',
    'Tennessee': 'TN',
    'Texas': 'TX',
    'Utah': 'UT',
    'Vermont': 'VT',
    'Virgin Islands': 'VI',
    'Virginia': 'VA',
    'Washington': 'WA',
    'West Virginia': 'WV',
    'Wisconsin': 'WI',
    'Wyoming': 'WY'
}
vaccine_data_state['location'] = vaccine_data_state['location'].map(us_state_abbrev)
vaccine_data_state = vaccine_data_state.fillna(0)
vaccine_data_state.head()
```

```python
# Load country vaccine data
vaccine_data_us = pd.read_csv('us_vacc.csv')
print(len(vaccine_data_us))
vaccine_data_us.head()
```

```python
# Clean country vaccine data
vaccine_data_us = vaccine_data_us.drop(['source_url'], axis=1)
vaccine_data_us = vaccine_data_us.drop(['location'], axis=1)
vaccine_data_us = vaccine_data_us.fillna(0)
vaccine_data_us.head()
```

```python
# exports cleaned datasets
county_data.to_csv('../cleaned_data/county_covid_clean.csv', index=False)
state_data.to_csv('../cleaned_data/state_covid_clean.csv', index=False)
us_data.to_csv('../cleaned_data/us_covid_clean.csv', index=False)
vaccine_data_state.to_csv('../cleaned_data/state_vaccine_clean.csv', index=False)
vaccine_data_us.to_csv('../cleaned_data/us_vaccine_clean.csv', index=False)
```
